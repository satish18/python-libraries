# Reshape Using Melt

    Melt is used to transform or reshape data

```python
import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1[df1['variable'] == 'chicago'] #selecting particular city
```

```python
df1 = pd.melt(df, id_vars=['day'], var_name='city', value_name='temperature')
df1
```

```python
#if you have multi-index columns
df.columns = [list('ABCD'), list('DEFG')]
df
```
