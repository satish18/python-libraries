# stack unstack

Dataframe reshaping technique stack .

Pandas stack method is used to transpose innermost level of columns in a dataframe. unstack() is used to perform a reverse operation.

```python
import pandas as pd
df = pd.read_excel('stocks.xlsx')
df
```

```python
df = pd.read_excel('stocks.xlsx', header=[0, 1])
df
```

```python
df.stack()
```

```python
df.stack(level=0)
```

```python
df_stack = df.stack()
df_stack
```

```python
df2 = pd.read_excel('stocks_3_levels.xlsx', header=[0, 1, 2])
df2
```

```python
df2.stack()
```

```python
df2.stack(level=0)
```

```python
df2.stack(level=1)
```

```python
df2.stack(level=2)
```

```python
df2.unstack()
```
